```python
# the ways to style using set_style():
# white (default), dark , whitegrid , darkgrid , ticks
# set_context() allows us to override default parameters . This affects things like size of the labels, 
# and other elements of your plot, but not the overall style -> set_style() . The contexts are:
# paper - small elements , ideal for printing in academic papers
# notebook (default) - medium size, good for working in jupyter notebook
# talk - larger text and lines , good for presentation slides where readability matters
# poster - Very large elements, suitable for large displays or posters
```

```python
sns.set_style('ticks')
sns.set_context('notebook')
sns.scatterplot(data = penguins , x = 'flipper_length_mm', y = 'body_mass_g', hue = 'species')
sns.despine(left = True, bottom = True) #top and right spines(borders) are removed along with left and bottom
```

```python
# for styling colors there's palletes 
# Set1 - bright and high contrast colors (9 distinct colors)
# Set2 - softer , more pastel like colors (8 distinct colors)
# Set3 - a mix of soft and bold , varied hues (12 distinct colors)
```

```python
# Palette will have different color parameters
# deep, muted , bright , pastel, dark, colorblind
```

Palette type     |                  Purpose                     |            Examples                |              Best used for 
Qualitative      | Categorical in nature , the color assigned to|  'Set1', 'Set2', 'Set3', 'Pastel1',|  Distinguishing categories or groups
                 | each group need to be distinct.              |   'Dark2', 'Colorblind', 'Accent'  |
Sequential       | Data with ordered progression (low -> high)  |   'Blues','Greens','Purples'       | Showinng maginitude or intensity like population
                 |                                              |   'OrRd','YiGnBu','BuPu','Greys'  | or revenue
Diverging        |Data with a meaningful midpoint(e.g., 0) or   | 'coolwarm','RdBu','BrBG','PiYG',   | Showing positive vs negative change (e.g.,profit/)
                 | mixed value like +ve and -ve (low to high)   | 'Spectral', 'terrain_r'            | loss, temperature change)

```python
# countplot
sns.countplot(data= penguins, x = 'species', hue ='island', palette='Set2') 
```

```python
# box plot used for detecting outliers that is extremely low or high values in column or feature, identifiying different groups 25%,50%,75%,100%,
# if there's a bubble on one extreme means that its a single data point
sns.boxplot(data= penguins, x = 'species', y = 'body_mass_g', hue='island') 
```

```python
sns.boxplot(data= penguins, x = 'species', y = 'body_mass_g', hue='sex') 
```

```python
# violin plot : similar to box plot but uses KDE(kernal density estimation) , whereever mid line ends thats the extreme
# density of data is width of the plot
# middle line median
# thick bar : interquartile range(if inner = 'box') (3rd quartile minus 1st quartile)
sns.violinplot(data= penguins, x = 'species', y = 'body_mass_g', hue='sex') 
```

```python
sns.violinplot(data= penguins, x = 'species', y = 'body_mass_g', hue='sex' , split=True) # it splits between male / female
```

```python
sns.violinplot(data= penguins, x = 'species', y = 'body_mass_g', palette = 'pastel') 
sns.swarmplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'island', size=3, palette ='dark')
#swarm fits exactly inside violin plot
```

```python
# kde(Kernal Density Estimation) plot : smooth version of Histogram
sns.kdeplot(data= penguins, x = 'body_mass_g',  hue = 'species', fill = True) 
# sns.kdeplot(data= penguins, x = 'body_mass_g',  hue = 'species', fill = False) 
```

```python
# Heatmap : a 2D color-coded matrix that shows relationships or patterns between two variables
# It's commonly used to:
# - Show correlation btw features
# - Visualise confusion matrices -> will see in the later stages
# - Highlight missing data -> sns.heatmap(df.isnull())
```

```python
columns = ['bill_length_mm', 'bill_depth_mm',
       'flipper_length_mm', 'body_mass_g']

penguins[columns].corr() #gives co relations among the fields
```

```python
sns.heatmap(data = penguins[columns].corr(), annot = True, cmap = 'Blues',vmin = -0.2, linewidth = 2, linecolor = 'black')
```

```python
# Rug plot : adds small ticks like rugs of datapoints used kde, histograms
sns.rugplot(data= penguins, x = 'body_mass_g', hue = 'species', palette = 'Set2', height=0.3)
```

```python
# pair - plot: shows relations btw all numericals , pair - wise, works only with numericals ones
sns.pairplot(data = penguins, hue = 'sex', palette = 'Set2', diag_kind = 'kde') #it can also be histogram , diag is only for these two
```

```python
#pair grid -> subplot grid for plotting pairwise relationships in a dataset.

graph = sns.PairGrid(data = penguins, hue = 'sex', palette = 'Set2')
graph.map_upper(sns.scatterplot) #upper_triangle
graph.map_lower(sns.kdeplot) #lower_triangle
graph.map_diag(sns.histplot) #diagonal
```
